# Observando los datos, Estacionariedad, Estacionalidad, Auto Correlacion

```python
import statsmodels.tsa.stattools as sts
```

Hipotesis nula = Es estacionaria.

Se observara si la hipotesis nula se rechaza o no, utilizando el valor estadistico y el valor critico

```python
sts.adfuller(data.Nacional)
```


En este caso. El valor estadistico es menor que el valor critico del 1%. Se rechaza la Hipotesis nula.\
Es decir.\
La serie temporal no es estacionaria

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

Observando si la serie temporal en los Casos a Nivel Nacional, tienen Estacionalidad.


```python
result = seasonal_decompose(data.Nacional,
                            model='additive',
                           )
result.plot()
plt.show()
```

# Observando el efecto de tendencia y ciclico

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
cycle_cases, trend_cases = hpfilter( data['Nacional'] ,lamb=1600)
```

```python
data['Trend'] = trend_cases
data.head(3)
```

```python
data[['Trend', 'Nacional']].plot(figsize=(12,8))
```

```python
data[['Trend', 'Nacional']]['2021-03-01':].plot(figsize=(12,8))
```

```python
import statsmodels.graphics.tsaplots as sgt
```

```python
sgt.plot_acf(data['Nacional'],
            lags=40,
            zero=False)

plt.title('ACF - Casos a nivel Nacional')
plt.show()
```
